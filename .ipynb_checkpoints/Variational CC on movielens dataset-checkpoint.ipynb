```python
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix, csc_matrix, csr_matrix
import matplotlib.pyplot as plt
%matplotlib inline

def plot_coo_matrix(m, size, m_size):
    if not isinstance(m, coo_matrix):
        m = coo_matrix(m)
    fig = plt.figure(figsize=size)
    ax = fig.add_subplot(111, axisbg='lightgray')
    ax.plot(m.col, m.row, 's', color='black', ms=m_size)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return (fig,ax)
```

```python
## Data format
## user id | item id | rating | timestamp
data = pd.read_csv('ml-100k/u1.base', sep='\t', header=None, engine='c')
data.columns = ['user', 'item', 'rating', 'timestamp']
data = coo_matrix((np.ones(len(data.rating)),(data.user-1,data.item-1))) #Binarization
```

```python
(fig, ax) = plot_coo_matrix(data, (15,10), 0.5)
ax.set_title("Original data")
ax.set_xlabel('movies')
ax.set_ylabel('users')
#plt.show(ax)
plt.savefig('./figures/movielensData.pdf', bbox_inches='tight')
```
